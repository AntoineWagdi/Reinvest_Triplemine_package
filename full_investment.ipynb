```python
import datetime
import calendar
import math
import pandas as pd
```

```python
x=int(input("Enter your package:"))
```

```python
today=datetime.date.today()
print(today)
p=int(x/50)
d=(x*0.005)
print("the duplication need :{} cycles".format(p))
print("your daily income will be:{}".format(d))
q=x*2+50
h=x+50
```

```python
package=[]
daily_income=[]
n=[]
after=[]
today_s=[]
```

```python
for i in range(h,q,50):
    package.append(i)
    y=(i*0.005)
    daily_income.append(y)
    z=(50/y)
    n.append(z)
```

```python
for c in n:
    a=math.ceil(c)
    after.append(a)
    today+=datetime.timedelta(days=a)
    today_s.append(today)
```

```python
g={'package':pd.Series(package),'daily_income':pd.Series(daily_income),'after':pd.Series(after),'On':pd.Series(today_s)}
```

```python
df = pd.DataFrame(g)
```
